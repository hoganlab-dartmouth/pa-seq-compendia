# Make figure 3

```python
from IPython.display import Image, display, SVG
import svgutils.transform as sg
import numpy as np
from lxml import etree
import os
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
```

### Function to plot

```python
def make_figure_panel(filename, scale_x_input, scale_y_input, x_loc, y_loc):
    panel = sg.fromfile(filename)

    panel_size = (
        np.round(float(panel.root.attrib["width"][:-2]) * 1.33, 0),
        np.round(float(panel.root.attrib["height"][:-2]) * 1.33, 0),
    )

    scale_x = scale_x_input
    scale_y = scale_y_input

    print(f"original: {panel_size}")
    print(f"scaled:{(panel_size[0]*scale_x,panel_size[1]*scale_y)}")

    panel = panel.getroot()
    panel.scale_xy(x=scale_x, y=scale_y)
    panel.moveto(x_loc, y_loc)

    return panel
```

```python
# Create panels for figure 3
panel_3a = make_figure_panel(
    "Expression_accessory_genes_all_samples_log10.svg",
    scale_x_input=0.8,
    scale_y_input=0.8,
    x_loc=10,
    y_loc=20,
)
panel_3b = make_figure_panel(
    "pa14_compendium.svg",
    scale_x_input=0.8,
    scale_y_input=0.8,
    x_loc=450,
    y_loc=20,
)
panel_3c = make_figure_panel(
    "pao1_compendium.svg",
    scale_x_input=0.8,
    scale_y_input=0.8,
    x_loc=890,
    y_loc=20,
)
panel_3d = make_figure_panel(
    "pa14_threshold_dist.svg",
    scale_x_input=0.8,
    scale_y_input=0.8,
    x_loc=10,
    y_loc=300,
)
panel_3e = make_figure_panel(
    "pao1_threshold_dist.svg",
    scale_x_input=0.8,
    scale_y_input=0.8,
    x_loc=450,
    y_loc=300,
)
```

```python
panel_3a_label = sg.TextElement(10, 20, "A", size=18, weight="bold", font="Verdana")
panel_3b_label = sg.TextElement(450, 20, "B", size=18, weight="bold", font="Verdana")
panel_3c_label = sg.TextElement(890, 20, "C", size=18, weight="bold", font="Verdana")
panel_3d_label = sg.TextElement(10, 300, "D", size=18, weight="bold", font="Verdana")
panel_3e_label = sg.TextElement(450, 300, "E", size=18, weight="bold", font="Verdana")
```

```python
figure_3 = sg.SVGFigure("1400", "600")
figure_3.append(
    [
        etree.Element("rect", {"width": "100%", "height": "100%", "fill": "white"}),
        panel_3a,
        panel_3b,
        panel_3c,
        panel_3d,
        panel_3e,
        panel_3a_label,
        panel_3b_label,
        panel_3c_label,
        panel_3d_label,
        panel_3e_label,
    ]
)
display(SVG(figure_3.to_str()))
```

```python
# save generated SVG files
figure_3.save("figure_3.svg")
```
